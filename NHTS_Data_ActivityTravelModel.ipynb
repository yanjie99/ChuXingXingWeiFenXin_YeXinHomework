```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('at.csv')
data.head()
```

```python
len(data)
```

```python
data3 = data
data3.head()
```

```python
data3['new_id'] = data3['per_id'] != data3['per_id'].shift()
data3['new_id'] = data3['new_id'].cumsum()
data3[:50]
# data3.to_csv('new_id.csv')
```

```python
data4 = data3.drop(['hh_id','per_id'], axis=1)
data4[:50]
```

```python
data4['student'] = data4['ActType'] == 8
# 学生
data5 = data4[data4['student']]
print(len(data5['new_id']))
data5['lasttime'] = data5['EndTime'] - data5['BegTime']
# data5.drop(['at_id', 'act_id', 'trip_id', 'O_pur'])
x1 = set(data5['new_id'])
len(x1)
```

```python
# worker
data4['worker'] = (data4['ActType'] == 2)|(data4['ActType'] == 3)|(data4['ActType'] == 4)|(data4['ActType'] == 5)
data6 = data4[data4['worker']]
x2 = set(data6['new_id'])
len(x2)
```

```python
# worker和student 的ID号
x = []
for i in x1:
    x.append(i)
for i in x2:
    x.append(i)
x = set(x)
print(len(x))
```

```python
# 总人数 ID个数
x3 = set(data4['new_id'])
print(len(x3))
# Non_worker 的ID号
x4 = x3 - x
print(len(x4))
print(x4)
```

```python
# data4 = data4.drop(['worker','student','new_id_2'], axis=1)
```

```python
# data4['new_id_2'] = data4['new_id']
dt1 = data4.groupby('new_id')
```

```python
x1
```

```python
worker = []
student = []
non_worker = []

for x in dt1:
    if x[0] in x4 :
        non_worker.append(x[1])
    elif x[0] in x1:
        student.append(x[1])
    elif x[0] in x2:
        worker.append(x[1])
    else:
        pass
#     print(x[0])
#     print(list(x))
```

```python
data_student = pd.concat(student)
data_worker = pd.concat(worker)
data_nonworker = pd.concat(non_worker)
```

```python
data_student.to_csv('data_student.csv')
data_worker.to_csv('data_worker.csv')
data_nonworker.to_csv('data_nonworker.csv')
```
